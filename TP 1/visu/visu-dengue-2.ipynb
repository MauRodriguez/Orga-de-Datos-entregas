# VISU - DENGUE -  Box Plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dengue_features = pd.read_csv("Dengue/dengue_features_train.csv")
dengue_labels = pd.read_csv("Dengue/dengue_labels_train.csv")
```

```python
merge = dengue_features.merge(dengue_labels, on=["city","year","weekofyear"],how='inner')
merge = merge.loc[:,['city','year','weekofyear','station_precip_mm','total_cases']]
merge
```

```python
merge["mes"] = np.floor(merge.weekofyear/13.3)+1
merge
```

```python
fig, ax = plt.subplots(dpi=170)
ax.set_title('Precipitaciones totales por cuarter del año',fontsize=12)
sns.boxplot(x=merge.mes, y=merge.station_precip_mm,palette="Blues")
ax.set_ylabel('Precipitaciones totales (mm)',fontsize=10)
ax.set_xlabel('Cuarter',fontsize=10)
ax.set_yscale('log')
```
