# 18. Divida la tierra en bloques de latitud y longitud de 5x5, ¿Cuál es el bloque con menos (o ninguna) referencias? (⭐⭐)

```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
import seaborn as sns
```

```python
geo_cols = ["gt_id","gt_lat","gt_lon"]
geo = pd.read_csv("Wikipedia dataset 2c2021/geo_tags.csv", usecols = geo_cols)
geo.head()
```

```python
#sería una matriz de 72 * 36 = 2592
# - 90 a 90   -180 a 180
```

```python
geo.gt_lon = (180 - geo.gt_lon) * (geo.gt_lon >= 180) + geo.gt_lon * (geo.gt_lon < 180)
geo["lat_box"] = np.floor(geo.gt_lat/5)
geo["lon_box"] = np.floor(geo.gt_lon/5)
```

```python
gb = geo.groupby(["lat_box","lon_box"]).count()
gb = gb["gt_id"]
```

```python
gb = gb.unstack()
```

```python
gb.fillna(0)
```
