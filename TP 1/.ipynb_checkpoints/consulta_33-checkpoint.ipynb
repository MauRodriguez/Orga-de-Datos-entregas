# 33. La primera discusión creada (⭐)

```python
import pandas as pd
import numpy as np
```

```python
logs = pd.read_csv("Wikipedia dataset 2c2021/logs.csv", parse_dates = ["timestamp"])
logs
```

```python
logs_B = logs.loc[:,["item_id","timestamp","title"]]
```

```python
def foo(x):
    if x.find("Discusión:") == 0:
        return True
    return False
logs_B.title = logs_B.title.fillna(" ")
logs_B["discussion"] = logs_B["title"].transform(func = foo)
discussions = logs_B.loc[logs_B.discussion == True]
discussions
```

```python
resultado_final = discussions.nsmallest(1,"timestamp")
resultado_final
```
