```python
import pandas as pd
import numpy as np
```

```python
contents_cols = ["title","revision_timestamp"]
contents = pd.read_csv("Wikipedia dataset 2c2021/contents.csv", usecols = contents_cols, parse_dates = ["revision_timestamp"])
```

## 8.

```python
contents_B = contents
contents_B.title = contents_B.title.fillna(" ")
```

```python
contents_B["first_letter"] = contents_B["title"].transform(func = lambda x: x[0])
contents_B
```

```python
today = pd.Timestamp.utcnow()
today
```

```python
contents_B["antique_median"] = today - contents_B.revision_timestamp
contents_B
```

```python
contents_B.antique_median = contents_B.antique_median.transform(func = lambda x: x.total_seconds())
contents_B
```

```python
gb = contents_B.groupby("first_letter").agg({"antique_median":"median"})
gb
```

```python
gb.antique_median = gb.antique_median.transform(func = lambda x: today - pd.Timedelta(seconds=x))
```

```python
resultado_final = gb
resultado_final
```
