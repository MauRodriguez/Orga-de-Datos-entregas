```python
import pandas as pd
import numpy as np
```

```python
contents_cols = ["title"]
contents = pd.read_csv("Wikipedia dataset 2c2021/contents.csv", usecols = contents_cols)
```

```python
import re
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
    
```

```python
contents_B = contents
contents_B.title = contents_B.title.fillna(" ")
```

```python
contents_B["first_letter"] = contents_B["title"].transform(func = lambda x: x[0])
contents_B["last_letter"] = contents_B["title"].transform(func = lambda x: x[len(x) - 1])
contents_B
```

```python
def detect_emoji(x):
    if emoji_pattern.search(x):
        return True
    return False
contents_B["has_first"] = contents_B["first_letter"].transform(func = detect_emoji)
contents_B["has_last"] = contents_B["last_letter"].transform(func = detect_emoji)
contents_B
```

```python
resultado_final = contents_B.loc[(contents_B.has_first == True) | (contents_B.has_last == True) ,"title"]
resultado_final.to_frame()
```
