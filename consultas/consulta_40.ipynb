# 40. Para los contenidos geolocalizados: ¿Cúal es el contenido más cercano del que fue editado más recientemente? ¿Y la diferencia entre sus tiempos de edición? (⭐⭐⭐)

```python
import pandas as pd
import numpy as np
```

```python
geo_cols = ["gt_page_id","gt_lat","gt_lon"]
contents_cols = ["revision_timestamp","id"]

geo = pd.read_csv("Wikipedia dataset 2c2021/geo_tags.csv", usecols = geo_cols)
contents = pd.read_csv("Wikipedia dataset 2c2021/contents.csv", usecols = contents_cols, parse_dates = ["revision_timestamp"])
```

```python
geo_merge = geo.merge(contents,left_on = "gt_page_id", right_on = "id", how = "inner")
geo_merge
```

```python
most_recent = geo_merge.loc[geo_merge.revision_timestamp.nlargest(1).index,]
lat = most_recent.gt_lat.iloc[0]
lon = most_recent.gt_lon.iloc[0]
most_recent
```

```python
geo_merge.drop(index=0)
```

```python
geo_merge["distance_from_recent"] = np.sqrt((geo_merge.gt_lat-lat)**2 + (geo_merge.gt_lon - lon)**2)
geo_merge
```

```python
closest = geo_merge.loc[geo_merge.distance_from_recent.nsmallest(2).index,]
closest
```

```python
closest = closest.iloc[1]
closest
```

```python
diff = (closest.revision_timestamp.iloc[0] - closest.revision_timestamp.iloc[1])
diff_in_seconds = diff.total_seconds()
diff_in_seconds
```
